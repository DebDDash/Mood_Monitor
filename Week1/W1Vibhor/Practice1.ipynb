Bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is: Less than 3 –> ‘small’ 3-5 –> ‘medium’ ‘>=5 –> ‘large’

```python
pl = np.array(df[2])
new_pl = pl.astype(float)

# print(new_pl)
arr = []
for x in new_pl:
  if(x<3):
    arr.append(f'{x}: Small')
  elif(x>3 and x<5):
    arr.append(f'{x}: medium')
  else:
    arr.append(f'{x}: large')
# print(arr)
ps = np.array(arr)
print(ps)
```
